# Matplotlib

```python
import numpy
import matplotlib.pyplot as plt
plt.style.use("classic")
# https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html
```

# Matplotlib Line Spacing

```python
a=numpy.linspace(0,10,5)  #print for me 0 to 10 with interval of 5 units.
print(a)
```

```python
a=numpy.linspace(0,20,5)  #print for me 0 to 20 with interval of 5 units.
print(a)
```

# Plot Function

```python
a=numpy.linspace(0,10,100)
print(a)
plt.plot(a,a)
```

```python
a=numpy.linspace(0,10,100)
print(a)
plt.plot(a,numpy.sin(a))
```

```python
a=numpy.linspace(0,10,100)
print(a)
plt.plot(a,numpy.cos(a))  #cos function
```

```python
a=numpy.linspace(0,10,100)  #combined waveform
print(a)
plt.plot(a,numpy.cos(a))
plt.plot(a,numpy.sin(a))
```

# Matplotlib Inline Function

```python
a=numpy.linspace(0,10,100)
print(a)
plt.plot(a,numpy.sin(a),'-')
plt.plot(a,numpy.cos(a),'--')
```

# Subplot Function

```python
plt.figure()
plt.subplot(2,1,1)      #two rows, one column and first plot
a=numpy.linspace(0,10,100)
plt.plot(a,numpy.sin(a),'-')

plt.subplot(2,1,2)   #two rows, one column and second plot
plt.plot(a,numpy.cos(a),'--')
```

```python
plt.figure()
a=numpy.linspace(0,10,100)
plt.subplot(2,2,1)      #two rows, two columns and first plot
plt.plot(a,numpy.sin(a))
plt.title("sin curve")

plt.subplot(2,2,2)   #two rows, two columns and second plot
plt.plot(a,numpy.cos(a))
plt.title("cos curve")

plt.subplot(2,2,3)  #two rows, two columns and third plot
plt.plot(a,numpy.sinh(a))
plt.title("sin hyperbolic")

plt.subplot(2,2,4)   #two rows, two columns and fourth plot.
plt.plot(a,numpy.cosh(a))
plt.title("cos hyperbolic")
```

# Label Function

```python
a=numpy.linspace(0,10,100)   #label each functions.
print(a)
plt.plot(a,numpy.sin(a))
plt.title("A sine Curve")  
plt.xlabel(" a values or range")
plt.ylabel(" sine values of a")
```

# Color & Legend Function 

```python
a=numpy.linspace(0,10,100)   #label each functions.
print(a)
plt.plot(a,numpy.sin(a),'-y',label="sin wave") #"-" represents solid line "y represents yellow"
plt.plot(a,numpy.cos(a),':b',label="cos wave") #":" represents dotted line "b represents blue"
plt.legend()    #legend function.
```
